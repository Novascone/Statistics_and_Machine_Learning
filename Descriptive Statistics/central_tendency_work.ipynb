```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
N = 10001
n = 101
nbins = 30

dist1 = np.random.randn(n) - 1

y1, x1 = np.histogram(dist1, nbins)
x1 = (x1[1:]+x1[:-1])/2

y1[len(x1) - 1] += 1000
plt.plot(x1,y1,'b')
mean_dist1 = sum(dist1) / len(dist1)
plt.plot([mean_dist1,mean_dist1],[0,max(y1)], 'b--')
```

```python
# median

# log-normal distribution

shift   = 0
stretch = .7
n       = 100
nbins   = 10

# data

data = stretch*np.random.randn(n) + shift
data = np.exp(data)

# histogram

y,x = np.histogram(data,nbins)
x = (x[:-1]+x[1:])/2

x[len(x) - 1] += 20
# mean and median

data_mean = np.mean(data)
data_median = np.median(data)

# plot

plt.plot(x,y)
plt.plot([data_mean, data_mean],[0,max(y)], 'r--')
plt.plot([data_median, data_median],[0,max(y)], 'b--')
plt.title('Log-normal data histogram')
plt.show()
```
