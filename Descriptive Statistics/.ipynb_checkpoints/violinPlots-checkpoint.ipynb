```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

```python
## create data

n = 1000
thresh = 5

data = np.exp(np.random.rand(n))
data[data>thresh] = thresh + np.random.randn(sum(data>thresh)) * .1

# hist
plt.hist(data,30)
plt.title('Histogram')
plt.show()

# violin

plt.violinplot(data)
plt.title("Violin")
plt.show()
```
