```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
# create data distributions

N = 10001
nbins = 30

dist1 = np.random.randn(N)  - 1
dist2 = 3*np.random.randn(N)  
dist3 = np.random.randn(N)  + 1

y1, x1 = np.histogram(dist1, nbins)
x1 = (x1[1:]+x1[:-1])/2

y2, x2 = np.histogram(dist2, nbins)
x2 = (x2[1:]+x2[:-1])/2

y3, x3 = np.histogram(dist3, nbins)
x3 = (x3[1:]+x3[:-1])/2
```

```python
plt.plot(x1,y1,'b')
plt.plot(x2,y2,'r')
plt.plot(x3,y3,'k')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
# overlay means

mean_dist1 = sum(dist1) / len(dist1)
mean_dist2 = np.mean(dist2)
mean_dist3 = np.mean(dist3)

plt.plot(x1,y1,'b', x2,y2,'r', x3,y3,'k')
plt.plot([mean_dist1,mean_dist1],[0,max(y1)], 'b--')
plt.plot([mean_dist2,mean_dist2],[0,max(y2)], 'r--')
plt.plot([mean_dist3,mean_dist3],[0,max(y3)], 'k--')

plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
# "failure" scenario

# new distribution
dist4 = np.hstack((np.random.randn(N)-2, np.random.randn(N)+2))

# histogram

[y4, x4] = np.histogram(dist4, nbins)
x4 = (x4[:-1]+x4[1:])/2

# mean

mean_dist4 = np.mean(dist4)

plt.plot(x4,y4,'b')
plt.plot([mean_dist4,mean_dist4], [0,max(y4)], 'b--')


plt.xlabel('Data values')
plt.ylabel('Data counts')
plt.show()
```

```python
# median

# log-normal distribution

shift   = 0
stretch = .7
n       = 2000
nbins   = 50

# data

data = stretch*np.random.randn(n) + shift
data = np.exp(data)

# histogram

y,x = np.histogram(data,nbins)
x = (x[:-1]+x[1:])/2

# mean and median

data_mean = np.mean(data)
data_median = np.median(data)

# plot

fig,ax = plt.subplots(2,1,figsize=(4,6))
ax[0].plot(data,'.',color=[.5,.5,.5],label='Data')
ax[0].plot([1,n],[data_mean,data_mean],'r--',label='Mean')
ax[0].plot([1,n],[data_median,data_median],'b--',label='Median')
ax[0].legend()

ax[1].plot(x,y)
ax[1].plot([data_mean, data_mean],[0,max(y)], 'r--')
ax[1].plot([data_median, data_median],[0,max(y)], 'b--')
ax[1].set_title('Log-normal data histogram')
plt.show()
```

```python
# mode 

data = np.round(np.random.randn(10))

uniq_data = np.unique(data)
for i in range(len(uniq_data)):
    print(f'{uniq_data[i]} appears {sum(data==uniq_data[i])}times.')
    
print(' ')
print('The modal value is %g'%stats.mode(data)[0][0])
```
