```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## Gaussian

# number of discreteizations

N = 1001

x = np.linspace(-4,4,N)

gaussdist = stats.norm.pdf(x)

plt.plot(x,gaussdist)
plt.title('Analytic Gaussian (normal) Distribution')
# is this a probability distribution
# no. It doesn't sum to one
print(sum(gaussdist))
# how do we convert to a probability distribution?
```

```python
plt.title('Analytic Gaussian (normal) Probability Distribution')
probDist = gaussdist/sum(gaussdist)
plt.plot(x,probDist)
print(sum(probDist))

```

```python
# from a calculus perspective we want the area under this curve to be equal to one
# we can do this with a Riemanns sum
# we need the heights of the bars of the histogram which we have
# and the width of the bars of the histogram, which we don't have
# so what is the width of each bin at the x-axis?
# its just the difference in points of any successive values of x
# print(sum(gaussdist)) <---- height of bars
probForm = sum(gaussdist)*np.diff(x[:2]) # <---- multiply by the width of the bars 
# which is the difference in points, so we calculate the difference in points of
# the first two points
print(probForm) 
```

```python
## Normally-distributed random numbers

# parameters

stretch = 1 # vaiance sqrt of standard deviation
shift   = 5 # mean
n       = 1000

# create data
data = stretch*np.random.randn(n)

# plot data
plt.hist(data, 25)
plt.title('Empirical Normal Distribution')
plt.show()
```

```python
## Uniformially distributed numbers

# parameters 
stretch = 2 # not the variance
shift   = 0.5
n       = 10000

# create data
data = stretch*np.random.rand(n) + shift-stretch/2

# plot data
fig, ax = plt.subplots(2,1, figsize=(5,6))

ax[0].plot(data,'.',markersize=1)
ax[0].set_title('Uniform Data Values')

ax[1].hist(data,25)
ax[1].set_title('Uniform Data Histogram')

plt.show()
```

```python
## Log-normal Distribution

N = 1001
x = np.linspace(0,10,N)
lognormaldist = stats.lognorm.pdf(x,1)

plt.plot(x, lognormaldist)
plt.title('Analytic Normal Distribution')
plt.show()
```

```python
## Emperical log-normal distribution

stretch = 0.5 # equal to the standard deviation?
shift   = 5 # equal to the mean?
N       = 2000 # number of data points

# generate data
data = stretch*np.random.randn(N) + shift
data = np.exp(data)

# plot data

fig, ax = plt.subplots(2,1, figsize=(4,6))

ax[0].plot(data,'.')
ax[0].set_title('Log-normal Data Values')

ax[1].hist(data,25)
ax[1].set_title('Log-normal Data Histogram')

plt.show()
```

```python
## Binomial

# a binomial distribution is the probability of K heads in N coin tosses
# given a probability of p heads (e.g. , .5 is a fair coin)

n = 10 # number of coin tosses
p = .5 # probability

x = range(n+2)
bindist = stats.binom.pmf(x,n,p)

plt.bar(x,bindist)
plt.title('Binomial Distribution (n=%s, p=%g)'%(n,p))
plt.show()
```

```python
## t

x = np.linspace(-4,4,1001)
df = 200 # degrees of freedom
t = stats.t.pdf(x,df)

plt.plot(x,t)
plt.xlabel('t-value')
plt.ylabel('P(t | H$_0$)')
plt.title('t(%g) Distribution'% df)
plt.show()
# not a probability distribution need to convert like we did for the Gaussian
```

```python
## F

num_df  = 5 # degrees of freedom in numerator
dem_df = 100 # degrees of freedim in denominator

# values to evaluate
x = np.linspace(0, 10, 10001)

# the distribution
fdist = stats.f.pdf(x,num_df,dem_df)

plt.plot(x, fdist)
plt.title(f'F({num_df},{dem_df})Distribution')
plt.xlabel('F value')
plt.show()
```
