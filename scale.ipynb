```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def y(x):
    return x
```

```python
def ex(x):
    return np.exp(x)
```

```python
x = np.random.randint(100,size=1000)
data1 = np.zeros(len(x))
data2 = np.zeros(len(x))
for i in range(len(x)):
    data1[i] = y(x[i])
for j in range(len(x)):
    data2[j] = ex(x[j])


plt.plot(np.sort(data1))
plt.plot(np.sort(data2))
    
```
