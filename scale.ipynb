```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 11, 1)
y = x # identity function

print(x,y)
```

```python
ye = np.exp(x) # exponential function
print(ye)
```

```python


# plotting the two functions on a linear scale
plt.plot(x,y)
plt.plot(x,ye)
plt.title('Linear')
plt.show()
    
```

```python

# plotting the two functions on a logarithmic scale
plt.yscale('log')
plt.plot(x,y)
plt.plot(x,ye)
plt.title('Log')
plt.show()
```
