```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# number of data points
n = 10000

# data that shows a log normal distribution
data = np.exp( np.random.randn(n)/2)
plt.plot(data,'s')
```

```python
# number of bins 
k = 40

plt.hist(data, bins=k)
plt.show
```

```python
y,x = np.histogram(data, bins=k)

# bin centers
xx = (x[1:]+x[:-1])/2
print(y)
plt.plot(xx,y)
plt.show()
```

```python
proportions = []

for i in range(0, len(y)):
    proportions.append(100*(y[i]/n))
print(proportions)

plt.hist(proportions,bins=len(y))
```
